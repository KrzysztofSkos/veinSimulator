# Simulation parameters

- Ltency = 0 => ntetwork is synchronous
- Sigma (transmission slots) = 1
- Router - rectangle
- Frame length = 64us

```python
import pandas as pd
df_simp = pd.read_excel (r'nodeCountTT64_simplified1.xlsx')
df_norm = pd.read_excel (r'nodeCountTT64.xlsx')
```

```python
import numpy as np
data_simp = df_simp.to_numpy()
data_norm = df_norm.to_numpy()
```

```python
import matplotlib.pyplot as plt
plt.plot(data_simp[:, 0], data_simp[:, 1], data_norm[:, 0],  data_norm[:, 1])
plt.title("Nodes in simulated vein fragment")
plt.ylabel("Node count")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(data_simp[:, 0], data_simp[:, 1])
plt.title("Nodes in simulated vein fragment")
plt.ylabel("Node count (simplified)")
plt.grid()
plt.subplot(212)
plt.plot(data_norm[:, 0],  data_norm[:, 1])
plt.ylabel("Node count (normal)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
plt.plot(data_simp[:, 0], data_simp[:, 2]/100, data_norm[:, 0], data_norm[:, 2]/100)
plt.title("Broken frames due to collisions (mean from 100 simulations)")
plt.ylabel("Broken frames")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
plt.figure(2)
plt.subplot(211)
plt.plot(data_simp[:, 0], data_simp[:, 2])
plt.title("Broken frames due to collisions (mean from 100 simulations)")
plt.ylabel("Broken frames (simplified)")
plt.grid()
plt.subplot(212)
plt.plot(data_norm[:, 0],  data_norm[:, 2])
plt.ylabel("Broken frames (normal)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
plt.plot(data_simp[:, 0], data_simp[:, 3]/100, data_norm[:, 0], data_norm[:, 3]/100)
plt.title("Completed transmissions (mean from 100 simulations)")
plt.ylabel("Throughput [frames/s]")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
plt.figure(2)
plt.subplot(211)
plt.plot(data_simp[:, 0], data_simp[:, 3])
plt.title("Completed transmissions (mean from 100 simulations)")
plt.ylabel("Throughput (simplified)")
plt.grid()
plt.subplot(212)
plt.plot(data_norm[:, 0],  data_norm[:, 3])
plt.ylabel("Throughput (normal)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```
