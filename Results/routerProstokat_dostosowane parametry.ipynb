# Simulation parameters

- Ltency = 0 => ntetwork is synchronous
- Sigma (transmission slots) = 1
- Router - rectangle
- Frame length = 64us
- średnica żyły 9.08 mm

```python
import pandas as pd
df_base = pd.read_csv (r'nodeCountTT64_simp05_baza.csv')
df_simp = pd.read_csv (r'nodeCountTT64_simp05_2.csv')
df_gauss = pd.read_csv (r'nodeCountTT64_simp05_2.csv')
```

```python
df_base = df_base.apply(pd.to_numeric, errors='coerce')
df_simp = df_simp.apply(pd.to_numeric, errors='coerce')
df_gauss = df_gauss.apply(pd.to_numeric, errors='coerce')
```

```python
import numpy as np
data_base = df_base.to_numpy()
data_simp = df_simp.to_numpy()
data_gauss = df_gauss.to_numpy()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
plt.plot(data_base[:, 0], data_base[:, 1], data_simp[:, 0], data_simp[:, 1], data_gauss[:, 0], data_gauss[:, 1])
plt.title("Nodes in simulated vein fragment")
plt.ylabel("Node count")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
%matplotlib notebook
plt.figure(1)
plt.subplot(221)
plt.plot(data_base[:, 0], data_base[:, 1])
plt.title("Nodes in simulated vein fragment")
plt.ylabel("Node count")
plt.grid()
plt.subplot(222)
plt.plot(data_simp[:, 0],  data_simp[:, 1])
plt.ylabel("Node count (fcl)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.subplot(223)
plt.plot(data_gauss[:, 0],  data_gauss[:, 1])
plt.ylabel("Node count (gauss)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
%matplotlib notebook
plt.plot(data_base[:, 0], data_base[:, 2]/100, data_simp[:, 0], data_simp[:, 2]/100, data_gauss[:, 0], data_gauss[:, 2]/100)
plt.title("Broken frames due to collisions (mean from 100 simulations)")
plt.ylabel("Broken frames")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
%matplotlib notebook
plt.figure(2)
plt.subplot(221)
plt.plot(data_base[:, 0], data_base[:, 2])
plt.title("Broken frames due to collisions (mean from 100 simulations)")
plt.ylabel("Broken frames")
plt.grid()
plt.subplot(222)
plt.plot(data_simp[:, 0],  data_simp[:, 2])
plt.ylabel("Broken frames (fc)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.subplot(223)
plt.plot(data_gauss[:, 0],  data_gauss[:, 2])
plt.ylabel("Broken frames (pm)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
%matplotlib notebook
plt.plot(data_base[:, 0], data_base[:, 3]/100, data_simp[:, 0], data_simp[:, 3]/100, data_gauss[:, 0], data_gauss[:, 3]/100)
plt.title("Completed transmissions (mean from 100 simulations)")
plt.ylabel("Throughput [frames/s]")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```
