# Simulation parameters

- Ltency = 0 => ntetwork is synchronous
- Sigma (transmission slots) = 1
- Router - rectangle
- Frame length = 64us

```python
import pandas as pd
df_simp = pd.read_excel (r'nodeCountTT64_simplified1.xlsx')
df_fc = pd.read_excel (r'nodeCountTT64_simp1_floor_ceil.xlsx')
df_pm = pd.read_excel (r'nodeCountTT64_simp1_plusminus.xlsx')
df_gauss = pd.read_csv (r'nodeCountTT64_simp1_gauss1000_2.csv')
```

```python
df_gauss = df_gauss.apply(pd.to_numeric, errors='coerce')
```

```python
import numpy as np
data_simp = df_simp.to_numpy()
data_fc = df_fc.to_numpy()
data_pm = df_pm.to_numpy()
data_gauss = df_gauss.to_numpy()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
plt.plot(data_simp[:, 0], data_simp[:, 1], data_fc[:, 0],  data_fc[:, 1], data_pm[:, 0],  data_pm[:, 1], data_gauss[:, 0], data_gauss[:, 1])
plt.title("Nodes in simulated vein fragment")
plt.ylabel("Node count")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
%matplotlib notebook
plt.figure(1)
plt.subplot(221)
plt.plot(data_simp[:, 0], data_simp[:, 1])
plt.title("Nodes in simulated vein fragment")
plt.ylabel("Node count")
plt.grid()
plt.subplot(222)
plt.plot(data_fc[:, 0],  data_fc[:, 1])
plt.ylabel("Node count (fcl)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.subplot(223)
plt.plot(data_pm[:, 0],  data_pm[:, 1])
plt.ylabel("Node count (pm)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.subplot(224)
plt.plot(data_gauss[:, 0],  data_gauss[:, 1])
plt.ylabel("Node count (gauss)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
%matplotlib notebook
plt.plot(data_simp[:, 0], data_simp[:, 2]/100, data_fc[:, 0], data_fc[:, 2]/100, data_pm[:, 0], data_pm[:, 2]/100, data_gauss[:, 0], data_gauss[:, 2]/100)
plt.title("Broken frames due to collisions (mean from 100 simulations)")
plt.ylabel("Broken frames")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```
