# Wyjaśnienie oznaczeń
* bernoulli_right - poprawne dobranie parametrów rozkładu Bernoulliego i powtórzenie wyników z pracy dra Rafaela
* bernoulli_wrong - policzenie średniej liczby węzłów w obserwowanym fragmencie żyły z proporcji i dodanie liczby z rozkładu Bernoulliego + zaokrąglenie do liczby naturalnej. Wyniki o 10% zbyt duże
* gauss_right = gauss_wrong. Nie udało mi się powtórzyć wyników z publikacji używając rozkładu Gaussa.
* gauss_wrong - policzenie średniej liczby węzłów w obserwowanym fragmencie żyły z proporcji i dodanie liczby z rozkładu Gaussa + zaokrąglenie do liczby naturalnej. Wyniki o 10% zbyt duże

```python
import pandas as pd
df_bernoulli_right = pd.read_csv (r'nodeCount_bernoulli_right.csv')
df_bernoulli_wrong = pd.read_csv (r'nodeCount_bernoulli_wrong.csv')
df_gauss_right = pd.read_csv (r'nodeCount_gauss_wrong.csv')
df_gauss_wrong = pd.read_csv (r'nodeCount_gauss_wrong.csv')
```

```python
df_bernoulli_right = df_bernoulli_right.apply(pd.to_numeric, errors='coerce')
df_bernoulli_wrong = df_bernoulli_wrong.apply(pd.to_numeric, errors='coerce')
df_gauss_right = df_gauss_right.apply(pd.to_numeric, errors='coerce')
df_gauss_wrong = df_gauss_wrong.apply(pd.to_numeric, errors='coerce')
```

```python
import numpy as np
data_bernoulli_right = df_bernoulli_right.to_numpy()
data_bernoulli_wrong = df_bernoulli_wrong.to_numpy()
data_gauss_right = df_gauss_right.to_numpy()
data_gauss_wrong = df_gauss_wrong.to_numpy()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
plt.plot(data_bernoulli_right[:, 0], data_bernoulli_right[:, 1], data_bernoulli_wrong[:, 0], data_bernoulli_wrong[:, 1], data_gauss_right[:, 0], data_gauss_right[:, 1], data_gauss_wrong[:, 0], data_gauss_wrong[:, 1])
plt.title("Nodes in simulated vein fragment")
plt.ylabel("Node count")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.legend(["data_bernoulli_right", "data_bernoulli_wrong", "data_gauss_right", "data_gauss_wrong"])
```

```python
%matplotlib notebook
plt.figure(1)
plt.subplot(221)
plt.plot(data_bernoulli_right[:, 0], data_bernoulli_right[:, 1])
plt.title("Nodes in simulated vein fragment")
plt.xlabel("Total nodes in bloodstream")
plt.ylabel("Node count  (data_bernoulli_right)")
plt.grid()
plt.subplot(222)
plt.plot(data_bernoulli_wrong[:, 0],  data_bernoulli_wrong[:, 1])
plt.ylabel("Node count (data_bernoulli_wrong)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.subplot(223)
plt.plot(data_gauss_right[:, 0],  data_gauss_right[:, 1])
plt.ylabel("Node count (data_gauss_right)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.subplot(224)
plt.plot(data_gauss_wrong[:, 0],  data_gauss_wrong[:, 1])
plt.ylabel("Node count (data_gauss_wrong)")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
```

```python
%matplotlib notebook
plt.plot(data_bernoulli_right[:, 0], data_bernoulli_right[:, 2]/1000, data_bernoulli_wrong[:, 0],data_bernoulli_wrong[:, 2]/1000, data_gauss_right[:, 0], data_gauss_right[:, 2]/1000, data_gauss_wrong[:, 0], data_gauss_wrong[:, 2]/1000)
plt.title("Broken frames due to collisions (mean from 100 simulations)")
plt.ylabel("Broken frames")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.legend(["data_bernoulli_right", "data_bernoulli_wrong", "data_gauss_right", "data_gauss_wrong"])
```

```python
%matplotlib notebook
plt.plot(data_bernoulli_right[:, 0], data_bernoulli_right[:, 3]/1000, data_bernoulli_wrong[:, 0],data_bernoulli_wrong[:, 3]/1000, data_gauss_right[:, 0], data_gauss_right[:, 3]/1000, data_gauss_wrong[:, 0], data_gauss_wrong[:, 3]/1000)
plt.title("Completed transmissions (mean from 100 simulations)")
plt.ylabel("Throughput [frames/s]")
plt.xlabel("Total nodes in bloodstream")
plt.grid()
plt.legend(["data_bernoulli_right", "data_bernoulli_wrong", "data_gauss_right", "data_gauss_wrong"])
```
